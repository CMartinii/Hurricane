```python
import numpy as np
import seaborn as sns #this goes at the top of the file
import matplotlib.pyplot as plt
from fitting_functions import *

data = np.loadtxt("Rafael_Data.csv", delimiter=",", dtype=str, skiprows = 1)
#print(data)

time = [int(i) for i in data[:, 0]]  #gets the first column as strings
Pressure_mB = [int(i) for i in data[:, 1]]
wind_speed_kt = [int(i) for i in data[:, 2]]
category = [str(i) for i in data[:, 3]]

assert mB_to_Pa(5) == 500

assert round(kt_to_metric(150) ) == 77


Pressure_Pa = [mB_to_Pa(pres) for pres in Pressure_mB]
print(Pressure_Pa)

wind_speed_metric = [(kt_to_metric(speed)) for speed in wind_speed_kt]
print(wind_speed_metric)

ax = sns.scatterplot(x=time, y=Pressure_Pa, hue = category)
ax.set(xlabel="Time [Hours]", ylabel="Pressure [Pa]")
plt.title("Pressure vs Time")
plt.show()

ax = sns.scatterplot(x=time, y=wind_speed_metric, hue = category)
ax.set(xlabel="Time [Hours]", ylabel="Windspeed [m/s]")
plt.title("Wind Speed vs Time")
plt.show()
```
